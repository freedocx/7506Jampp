```python
import pandas as pd
import numpy as np
import unidecode
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import os
import seaborn as sns
import scipy.stats as stats
import matplotlib.colors as mcolors
import holoviews as hv
import bokeh
hv.notebook_extension('bokeh')
renderer = hv.renderer('bokeh')
import datetime as dt

pd.options.display.float_format = '{:20,.2f}'.format # suprimimos la notacion cientifica en los outputs

matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['font.family'] = 'Liberation Serif'
```

```python
data_types_clicks = {"advertiser_id" : 'category', "ref_type" : 'category',"timeToClick" : 'float16',\
                     "touchX" : 'float16','touchY':'float16', "brand" : 'category', "action_id" : 'category',\
                    "source_id" : 'category', "trans_id" : 'category','ref_hash':'category'
             }


data_types_installs = {"application_id" : 'category', "ref_type" : 'category',"click_hash" : 'category', "attributed" : bool ,\
                "implicit": bool, "device_countrycode" : 'category',"device_brand" : 'category', "device_model" : 'category', \
                 "session_user_agent" : 'category', "user_agent": 'category',"event_uuid" : 'category', "kind" : str, "wifi" : bool, \
                "trans_id" : 'category', "device_language" : 'category','ref_hash':'category'
             }
folder = os.getcwd()
```

```python
def autolabel(rects):
#### funcion que realiza etiquetas en los BARPLOTs
    for rect in rects:
        height = rect.get_height() #### obtiene la altura de la barra
        ax.text(rect.get_x() + rect.get_width()/2. #### posiciona el label horizontalmente
                , height, #### posiciona el label verticalmente
                '%d' % int(height),
                ha='center',
                va='bottom',
                fontsize= 14)
def removeSpines(ax):
#### remuevo los bordes, labels y lineas 
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.xaxis.set_ticks_position('none')
```

```python
clicks = pd.read_csv( folder + '/data/clicks.csv.gzip', compression='gzip',low_memory=False,\
                     parse_dates=['created'],infer_datetime_format=True, dtype=data_types_clicks)
installs = pd.read_csv( folder + '/data/installs.csv.gzip',parse_dates=['created'],infer_datetime_format=True,\
                     dtype=data_types_installs, compression='gzip',low_memory=False,\
                     )
```

```python
toplot = clicks[['ref_hash','advertiser_id']].groupby('ref_hash').agg('count')['advertiser_id'].sort_values(ascending=False).reset_index().head(10)
toplot.columns=['ref_hash','counts']
```

```python
tox=toplot.ref_hash.tolist()
height = toplot.counts.tolist()
fig, ax = plt.subplots(figsize=(8,5))
rects1 = ax.bar(tox, height , color='#1F85DE',edgecolor='#265680', align='center')
##############################################################################
#### titulos
ax.set_title('Cantidad de Clicks por Usuario', fontweight='bold')
ax.set_ylabel('Cantidad de Clicks', fontweight='bold')
ax.set_xlabel('Usuario', fontweight='bold')
ax.set_yticks([])
plt.xticks(rotation=60)
#####################################################################################
autolabel(rects1)
removeSpines(ax)
### aplico la funcion autolabel para poner la cantidad sobre cada barra
plt.savefig('CantidadDeClicksPorUsuario.png')
plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize=(16,9))
fig.suptitle('Distribucion de Clicks Banner1',fontsize = 25)
axes[0, 0].set_title(r'$(\gamma=%1.1f)$ 1)')
axes[0, 0].hist2d(banner1.touchX.tolist(), banner1.touchY.tolist(), bins=(300,300),\
                 cmap = 'coolwarm')
gammas = [0.8, 0.6, 0.4]
for ax, gamma in zip(axes.flat[1:], gammas):
    ax.set_title(r'$(\gamma=%1.1f)$' % gamma)
    ax.hist2d(banner1.touchX.tolist(),banner1.touchY.tolist(),cmap = 'coolwarm',\
              bins=(100,100), norm=mcolors.PowerNorm(gamma))
plt.savefig('ClicksEnPantallaBanner1.png')
plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize=(16,9))
fig.suptitle('Distribucion de Clicks Banner2',fontsize = 25)
axes[0, 0].set_title(r'$(\gamma=%1.1f)$ 1)')
axes[0, 0].hist2d(banner2.touchX.tolist(), banner2.touchY.tolist(), bins=(300,300),\
                 cmap = 'coolwarm')
gammas = [0.8, 0.6, 0.4]
for ax, gamma in zip(axes.flat[1:], gammas):
    ax.set_title(r'$(\gamma=%1.1f)$' % gamma)
    ax.hist2d(banner2.touchX.tolist(),banner2.touchY.tolist(),cmap = 'coolwarm',\
              bins=(100,100), norm=mcolors.PowerNorm(gamma))
plt.savefig('ClicksEnPantallaBanner2.png')
plt.show()
```
